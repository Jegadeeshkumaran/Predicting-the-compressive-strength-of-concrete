```python
import pandas as pd
import numpy as np
```

```python
concrete=pd.read_csv("Concrete Compressive Strength.csv")
concrete.head()
```

```python
concrete.rename(columns={'Cement (component 1)(kg in a m^3 mixture)': "cement",
 'Blast Furnace Slag (component 2)(kg in a m^3 mixture)': "blast",
 'Fly Ash (component 3)(kg in a m^3 mixture)': "fly_ash",
 'Water  (component 4)(kg in a m^3 mixture)': "water",
 'Superplasticizer (component 5)(kg in a m^3 mixture)': "superplasticizer",
 'Coarse Aggregate  (component 6)(kg in a m^3 mixture)': "coarse_aggregate",
 'Fine Aggregate (component 7)(kg in a m^3 mixture)': "fine_aggregate",
 'Age (day)':"age",
 'Concrete compressive strength(MPa, megapascals) ':"concrete_compressive"},inplace=True)
```

```python
concrete
```

```python
concrete.isnull().sum()
```

```python
print(f'The total number of missing values within the dataset is zero there are no null values as \n{concrete.isnull().sum()}')
```

```python
concrete_scaled=concrete.drop(["concrete_compressive"],axis=1)
concrete_scaled
```

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
```

```python
concrete_scaled=scaler.fit_transform(concrete_scaled)
```

```python
concrete_scaled
```

```python
x=concrete_scaled
y=concrete["concrete_compressive"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
lin_model=LinearRegression().fit(x_train,y_train)
```

```python
y_pred=lin_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,y_pred))
```
